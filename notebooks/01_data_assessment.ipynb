### Data Quality Assessment

```python
for name, df in dfs.items():
    quality_view = pd.DataFrame({
        "column": df.columns,
        "data_type": df.dtypes.astype(str).values,
        "non_null_count": df.notna().sum().values,
        "null_count": df.isna().sum().values,
        "empty_string_count": [
            df[col].astype(str).str.strip().eq("").sum() if df[col].dtype == "object" else 0
            for col in df.columns
        ]
    })

    display(pd.DataFrame({"dataset": [name]}))
    display(quality_view)
```

### Date Conversion

Date columns are initially loaded as text from CSV files. In this step, date fields are identified by name and converted to pandas datetime so date ranges and time-based patterns can be assessed correctly.

```python
date_columns = {
    "orders": [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ],
    "order_items": ["shipping_limit_date"],
    "reviews": ["review_creation_date", "review_answer_timestamp"]
}

date_conversion_summary = []

for dataset, columns in date_columns.items():
    for column in columns:
        if column in dfs[dataset].columns:
            null_before = dfs[dataset][column].isna().sum()

            dfs[dataset][column] = pd.to_datetime(dfs[dataset][column], errors="coerce")

            null_after = dfs[dataset][column].isna().sum()
            coerced_to_null = null_after - null_before

            date_conversion_summary.append({
                "dataset": dataset,
                "column": column,
                "null_before_conversion": null_before,
                "coerced_to_null": coerced_to_null,
                "converted_count": dfs[dataset][column].notna().sum()
            })

date_conversion_summary_df = pd.DataFrame(date_conversion_summary)

date_conversion_summary_df
```

> **Note:** The date conversion logic above has been extracted into `python/utils/date_conversion.py` as a reusable utility function `convert_dates(dfs)`. All downstream notebooks import and call this function directly rather than repeating the conversion logic, ensuring date columns are consistently handled across the entire notebook series.

```python
date_range_summary = []

all_min_dates = []
all_max_dates = []

for dataset, columns in date_columns.items():
    for column in columns:
        if column in dfs[dataset].columns:
            min_date = dfs[dataset][column].min()
            max_date = dfs[dataset][column].max()

            date_range_summary.append({
                "dataset": dataset,
                "column": column,
                "min_date": min_date.date() if pd.notna(min_date) else None,
                "max_date": max_date.date() if pd.notna(max_date) else None,
                "missing_values": dfs[dataset][column].isna().sum()
            })

            if pd.notna(min_date):
                all_min_dates.append(min_date)

            if pd.notna(max_date):
                all_max_dates.append(max_date)

date_range_summary_df = pd.DataFrame(date_range_summary)

display(date_range_summary_df)

overall_min_date = min(all_min_dates).date()
overall_max_date = max(all_max_dates).date()

print(f"Overall dataset date range: {overall_min_date} to {overall_max_date}")
```

> **⚑ Quality Flag — `shipping_limit_date` anomaly:** `shipping_limit_date` extends to 2020, approximately two years past the dataset's last order date (Oct 2018). Since this column represents a seller's handling deadline set at order time, a date this far out is unexpected. Investigated in `02_data_quality_check.ipynb` → *Investigating shipping_limit_date anomaly*.
